```python
Parameters in Dropna method:
1. axis: "rows" or  "columns"
2. how : "any" or "all"
3. subset:
4. thresh
```

```python
df
```

```python
# dropping columns having missing values in any of the 4 rows, but only drop it if it has less than 2 non-null values
df.dropna(axis="columns",how="any",thresh=2)
```

```python
df
```

```python
# dropping rows having missing values in any of the 4 columns, but only drop it if it has less than 1 non-null values
df.dropna(axis="rows",how="any",thresh=1)
```

```python
# dropping rows having missing values in any of the 4 columns, but only drop it if it has less than 2 non-null values
df.dropna(axis="rows",how="any",thresh=2)
```
