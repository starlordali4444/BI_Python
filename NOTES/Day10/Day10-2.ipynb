```python
plt.plot(matches,runs)
plt.show()
```

```python
plt.plot(matches,runs)
plt.xlabel("No. of matches Played")
plt.show()
```

```python
plt.plot(matches,runs)
plt.xlabel("No. of matches Played")
plt.ylabel("Runs Scored")
plt.title("Performance of Kohli in IPL 2021")
plt.show()
```

```python
plt.plot(matches,runs)
plt.xlabel("No. of matches Played")
plt.ylabel("Runs Scored")
plt.title("Performance of Kohli in IPL 2021")
plt.xticks([1,2,3,4])
plt.show()
```

```python
plt.plot(matches,runs)
plt.xlabel("No. of matches Played")
plt.ylabel("Runs Scored")
plt.title("Performance of Kohli in IPL 2021")
plt.xticks([1,2,3,4])
plt.yticks([0,50,100,150])
plt.show()
```
